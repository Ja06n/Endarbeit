```python
# Importiere die erforderlichen Bibliotheken
import turtle  # Importiere die Turtle-Grafikbibliothek
import math    # Importiere die Mathematikbibliothek

# Initialisiere Turtle
wn = turtle.Screen()  # Erstelle ein Fenster für die Turtle-Grafik
wn.setup(1000, 800)    # Setze die Größe des Fensters auf 1000x800 Pixel
myTurtle = turtle.Turtle()  # Erstelle ein Turtle-Objekt für das Zeichnen
myTurtle.pensize(3)   # Setze die Stiftgröße auf 3 Pixel

# Hauptprogramm
def main():
    valueOne = 0
    valueTwo = 1
    fib = 1
    for i in range(8):
        myTurtle.right(90)  # Drehe die Turtle um 90 Grad nach rechts
        drawSq(fib * 20)   # Zeichne ein Quadrat mit Seitenlänge entsprechend der Fibonacci-Zahl
        fib = valueOne + valueTwo
        valueOne = valueTwo
        valueTwo = fib

# Funktion zum Zeichnen eines Quadrats
def drawSq(seitenlaenge):
    for n in range(4):
        myTurtle.forward(seitenlaenge)  # Bewege die Turtle vorwärts um die Seitenlänge
        myTurtle.left(90)   # Drehe die Turtle um 90 Grad nach links

# Funktion für die Fibonacci-Quadrate
def sprial():
    myTurtle.color("red")  # Setze die Stiftfarbe auf Rot
    myTurtle.speed(2)      # Setze die Zeichengeschwindigkeit auf 2 (mittlere Geschwindigkeit)
    r = 20
    angle = 90
    myTurtle.right(90)     # Drehe die Turtle um 90 Grad nach rechts
    myTurtle.penup()       # Hebe den Stift an, um ohne zu zeichnen zu bewegen
    myTurtle.setpos(0, 0)  # Setze die Turtle-Position auf (0, 0)
    myTurtle.pendown()     # Senke den Stift, um zu zeichnen
    arc(20, angle)         # Zeichne einen Bogen mit Radius 20 und Winkel 90 Grad
    arc(20, angle)
    arc(40, angle)
    arc(60, angle)
    arc(100, angle)
    arc(160, angle)
    arc(260, angle)
    arc(420, angle)

# Funktion zum Zeichnen einer Linie aus mehreren Teilen
def arcLine(n, laenge, winkel):
    for i in range(n):
        myTurtle.forward(laenge)  # Bewege die Turtle vorwärts um die Länge
        myTurtle.left(winkel)     # Drehe die Turtle um den Winkel nach links

# Funktion zum Zeichnen eines Bogens
def arc(r, winkel):
    bogenlaenge = 2 * math.pi * r * abs(winkel) / 360  # Berechne die Bogenlänge
    n = int(bogenlaenge / 4) + 1  # Berechne die Anzahl der Abschnitte
    schrittlaenge = bogenlaenge / n
    schritt_winkel = float(winkel) / n
    myTurtle.left(schritt_winkel / 2)
    arcLine(n, schrittlaenge, schritt_winkel)
    myTurtle.right(schritt_winkel / 2)

# Hauptprogrammschleife
main()   # Rufe die Hauptfunktion auf, um die Fibonacci-Quadrate zu zeichnen
sprial() # Rufe die Funktion auf, um die rote Spirale zu zeichnen
wn.exitonclick()  # Schließe das Turtle-Fenster, wenn es angeklickt wird
```
